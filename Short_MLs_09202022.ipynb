### **Case 1. 2 Circle Data Classification** ###

```python
import tensorflow as tf
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
%matplotlib inline

print(f'TensorFlow: {tf.__version__}')
print(f'Keras     : {tf.keras.__version__}')
```

```python
# make data: two circles on x-y plane as a classification problem
# about make_circles(), https://scikit-learn.org/0.24/modules/generated/sklearn.datasets.make_circles.html#sklearn.datasets.make_circles
X, y = make_circles(n_samples=1000, factor=0.5, noise=0.1)
plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=y)
plt.colorbar()
plt.xlabel('$X0$', fontsize=18)
plt.ylabel('$X1$', fontsize=18)
# plt.legend(loc="upper right", fontsize=18)
plt.show()
```

```python
# input layer   : X(dim=2), y(dim=1)
# hidden layer 1: 5,      activation func: relu
# output layer  : 5 to 1, activation func(hypothesis): sigmoid
# optimizer: adam, loss func: binary cross entropy(BCE), metrics(scores): accuracy

model = tf.keras.Sequential([                       # about tf.keras.Sequential(), https://runebook.dev/ko/docs/tensorflow/keras/sequential
    tf.keras.layers.InputLayer(input_shape=(2,)),   # about tf.keras.InputLayer(), https://runebook.dev/ko/docs/tensorflow/keras/layers/inputlayer
    tf.keras.layers.Dense(5, "relu"),               # about tf.keras.Dense(), https://runebook.dev/ko/docs/tensorflow/keras/layers/dense
    tf.keras.layers.Dense(1, "sigmoid")
])
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["acc"])

model.fit(X, y, batch_size=32, epochs=100, verbose=0)
```

```python
print(model.summary())
```

```python
# about model.evaluate(), https://keras.io/api/models/model_training_apis/#evaluate-method
print(model.evaluate(X,y))
print('[\tLoss\t\t    Accuracy\t]')
```

```python
# input layer   : X(dim=2), y(dim=1)
# hidden layer 1: 5,      activation func: sigmoid
# hidden layer 2: 5,      activation func: sigmoid
# hidden layer 3: 5,      activation func: sigmoid
# output layer  : 5 to 1, activation func(hypothesis): sigmoid
# optimizer: adam, loss func: binary cross entropy(BCE), metrics(scores): accuracy

model = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=(2,)),
    tf.keras.layers.Dense(5, "sigmoid"),
    tf.keras.layers.Dense(5, "sigmoid"),
    tf.keras.layers.Dense(5, "sigmoid"),
    tf.keras.layers.Dense(1, "sigmoid")
])
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["acc"])

model.fit(X, y, batch_size=32, epochs=100, verbose=0)
```

```python
print(model.summary())
```

```python
print(model.evaluate(X,y))
print('[\tLoss\t\t    Accuracy\t]')
```

```python
# input layer   : X(dim=2), y(dim=1)
# hidden layer 1: 5,      activation func: relu
# hidden layer 2: 5,      activation func: relu
# hidden layer 3: 5,      activation func: relu
# output layer  : 5 to 1, activation func(hypothesis): sigmoid
# optimizer: adam, loss func: binary cross entropy(BCE), metrics(scores): accuracy

model = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=(2,)),
    tf.keras.layers.Dense(5, "relu"),
    tf.keras.layers.Dense(5, "relu"),
    tf.keras.layers.Dense(5, "relu"),
    tf.keras.layers.Dense(1, "sigmoid")
])
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["acc"])

model.fit(X, y, batch_size=32, epochs=100, verbose=0)
```

```python
print(model.summary())
```

```python
print(model.evaluate(X,y))
print('[\tLoss\t\t    Accuracy\t]')
```
